### Parent Functions
<br></br>
<code>f(x) = x</code>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, num=100)

fx = []

for i in range(len(x)):
    fx.append(x[i])
    
plt.plot(x,fx)
plt.grid()
plt.axvline()
plt.axhline()
plt.show()
```

<code>f(x) = x^2</code>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, num=100)

fx = []

for i in range(len(x)):
    fx.append(x[i]**2)
    
plt.plot(x,fx)
plt.grid()
plt.axvline()
plt.axhline()
plt.show()
```

<code>f(x) = 1/x</code>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, num=100)

fx = []

for i in range(len(x)):
    fx.append(1 / x[i])
    
plt.plot(x,fx)
plt.grid()
plt.axvline()
plt.axhline()
plt.show()
```

<code>f(x) = |x|</code>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, num=100)

fx = []

for i in range(len(x)):
    fx.append(abs(x[i]))
    
plt.plot(x,fx)
plt.grid()
plt.axvline()
plt.axhline()
plt.show()
```
